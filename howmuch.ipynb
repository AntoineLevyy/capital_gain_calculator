```python
import os


import requests
from dotenv import load_dotenv
import matplotlib.pyplot as plt
import pandas as pd

key = os.getenv("AlphaVantageAPI_KEY")

symbol = "NVDA"
entry_date = "2010-08-31"
exit_date = "2023-09-15"
investment = 10000


# replace the key apikey below with your own key from https://www.alphavantage.co/support/#api-key
url = f'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol={symbol}&apikey=key'
r = requests.get(url)
data = r.json()

# Extract time series data
time_series_data = data['Time Series (Daily)']
# Extract timestamps and close prices
timestamps = list(time_series_data.keys())[::-1]
close_prices = [float(time_series_data[time]['4. close']) for time in timestamps]
```

```python

#calculating the capital_gain between entry and exit:
def capital_gain(entry_date, exit_date, investment):
    gain = (float(time_series_data[exit_date]["4. close"])/float(time_series_data[entry_date]["4. close"]))*investment
    return gain


capital_gain(entry_date,exit_date,investment)
```

```python
# Plotting the data
plt.figure(figsize=(15, 7))
plt.plot(timestamps, close_prices, marker='o', linestyle='-', color='b')
plt.title(f'{symbol} Daily Prices')
plt.xlabel('Timestamp')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()


# Display the plot
plt.show()

```

```python
import os


import requests
from dotenv import load_dotenv
import matplotlib.pyplot as plt

key = os.getenv("AlphaVantageAPI_KEY")

# replace the key apikey below with your own key from https://www.alphavantage.co/support/#api-key
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=NVDA&apikey=key'
r = requests.get(url)
data = r.json()
```

```python

# Extract time series data
time_series_data = data['Time Series (Daily)']
# Extract timestamps and close prices
timestamps = list(time_series_data.keys())[::-1]
close_prices = [float(time_series_data[time]['4. close']) for time in timestamps]
# Plotting the data
plt.figure(figsize=(15, 7))
plt.plot(timestamps, close_prices, marker='o', linestyle='-', color='b')
plt.title('GOOG Daily Prices')
plt.xlabel('Timestamp')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

# Display the plot
plt.show()

```

```python
def capital_gain(entry_date, exit_date, investment):
    gain = (float(time_series_data[exit_date]["4. close"])/float(time_series_data[entry_date]["4. close"]))*investment
    return gain
```
